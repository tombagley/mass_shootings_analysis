```python
#drop all firearms not used in shooting
firearms = firearms[firearms['Used in Shooting?'] != 0]
firearms = firearms.reset_index()
#fill all missing data with the mode which is zero for all variables
firearms = firearms.fillna(0)
firearms = firearms.drop(columns=['index'])
```

```python
#this function takes a data set and searches each variable for the max and creates a new data point with the max filled in for each variable
def combine_rows(df):
  c = df['Classification']
  c_list = []
  for i in range(len(c)):
    c_list.append(c[i])
  c = max(c_list)

  ca = df['Caliber']
  ca_list = []
  for i in range(len(ca)):
    ca_list.append(ca[i])
  ca = max(ca)

  m = df['Modified']
  m_list = []
  for i in range(len(m)):
    m_list.append(m[i])
  m = max(m_list)

  lcm = df['Large Capacity Magazine']
  lcm_list = []
  for i in range(len(lcm)):
    lcm_list.append(lcm[i])
  lcm = max(lcm_list)

  em = df['Extended Magazine']
  em_list = []
  for i in range(len(em)):
    em_list.append(em[i])
  em = max(em_list)

  ip = df['Illegal Purchase']
  ip_list = []
  for i in range(len(ip)):
    ip_list.append(ip[i])
  if max(ip_list) > 0:
    ip = 1
  else:
    ip = 0

  data = [[df.iloc[0, 0], df.iloc[0, 1], df.iloc[0, 2], c, ca, m, lcm, em, ip]]
  df = pd.DataFrame(data, columns=['Case #', 'Shooter Last Name', 'Shooter First Name', 'Classification', 'Caliber', 'Modified', 'Large Capacity Magazine', 'Extended Magazine', 'Illegal Purchase'])
  return df
```

```python
#This function does the same thing as the previous although this one is specifically used for spliting the incidences where multiple shooters where in the same shooting. This creates a copy of each data point or one for each shooter.
def combine_rows_case(df, casenum):
  c = df['Classification']
  c_list = []
  for i in range(len(c)):
    c_list.append(c[i])
  c = max(c_list)

  ca = df['Caliber']
  ca_list = []
  for i in range(len(ca)):
    ca_list.append(ca[i])
  ca = max(ca)

  m = df['Modified']
  m_list = []
  for i in range(len(m)):
    m_list.append(m[i])
  m = max(m_list)

  lcm = df['Large Capacity Magazine']
  lcm_list = []
  for i in range(len(lcm)):
    lcm_list.append(lcm[i])
  lcm = max(lcm_list)

  em = df['Extended Magazine']
  em_list = []
  for i in range(len(em)):
    em_list.append(em[i])
  em = max(em_list)

  ip = df['Illegal Purchase']
  ip_list = []
  for i in range(len(ip)):
    ip_list.append(ip[i])
  if max(ip_list) > 0:
    ip = 1
  else:
    ip = 0

  data = [[casenum, df.iloc[0, 1], df.iloc[0, 2], c, ca, m, lcm, em, ip]]
  df = pd.DataFrame(data, columns=['Case #', 'Shooter Last Name', 'Shooter First Name', 'Classification', 'Caliber', 'Modified', 'Large Capacity Magazine', 'Extended Magazine', 'Illegal Purchase'])
  return df
```

```python
#start the new data set with the first row
first_row = firearms[firearms['Case #'] == '1']
df = combine_rows(first_row)
```
