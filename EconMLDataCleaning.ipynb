# There are many many missing values that we will need to consider


```python
#Code from Parker that counts how many nulls are in each column
missingVals = data.isnull().sum().sort_values(ascending=False)
print(missingVals[:20]) #Print out columns missing the most data
dropCols = missingVals[missingVals > 15]
notEnoughVals = list(dropCols.index) #Drop all columns that have more than 15 missing values
notEnoughVals
```

### Drop Columns with lots of missing data 
### Drop columns that are strings unrelated/unhelpful to the analysis that can't be made into useful dummies

```python
#All of these columns are strings that are not relevant to the full dataset. We will drop these.
stringCols = ['Street Number','Street Name','Specify Other Weapons or Gear','County','Shooter First Name','Shooter Last Name','State Code','Full Date','Zip Code','School Performance Specified','Community Involvement','Signs of Crisis Expanded','Psychiatric Medication','Health Issues - Specify','Specify Relationship to Other Shooting(s)','Specify Pop Culture Connection','Who Killed Shooter On Scene','City']

#Combine with the columns with missing data and drop them
dropTheseCols = stringCols + notEnoughVals
dfNew = data.drop(dropTheseCols,axis=1)

print(f"Num Cols: {len(dfNew.columns)}")
```

```python
dfNew = pd.get_dummies(dfNew) #We don't need to drop the first column of dummies because there are missing values that won't be made into dummies
dfNew = pd.get_dummies(dfNew,columns=['Location','Children','Criminal Sentence','Highest Level of Justice System Involvement','Known Hate Group or Chat Room Affiliation','Region',
                            'Gender','Relationship Status','History of Domestic Abuse','Violent Video Games','Motive: Religious Hate','Role of Psychosis in the Shooting',
                            'Firearm Proficiency',
                            'On-Scene Outcome',
                            'Insanity Defense',
                            'Urban/Suburban/Rural',
                            'Specify Armed Person',
                            'Military Service',
                            'History of Physical Altercations',
                            'Suicidality',
                            'Voluntary or Involuntary Hospitalization',
                            'Motive: Racism/Xenophobia',
                            dfNew.columns[95],dfNew.columns[98], #This is hardcoded to put in Motive: Other and Social Media Use since the strings weren't working
                            'Pop Culture Connection'])
dfNew.head()
```

### Fill in Missing Values
Now that we've accounted for columns with too many missing values by dropping them, and added in dummy variables, we can fill in the rest of the values with either the mean or the mode. 

```python
#How many are missing
print(dfNew.isnull().sum().sort_values(ascending=False))

def find_mode(column): #This function is from the internet to 
  vals, counts = np.unique (column, return_counts=True)
  index = np.argmax (counts)
  return (vals[index])

missingVals = dfNew.isnull().sum().sort_values(ascending=False)
colsWithNull = missingVals[missingVals > 0].index

#Fill with mean value
for col in colsWithNull:
  print(col)
  if max(dfNew[col]) <= 9: #This should capture any categorical variables we missed in our one-hot encoding
    #Fill with mode
    modeVal = find_mode(dfNew[col].to_numpy()) #Find mode of columns
    dfNew[col] = dfNew[col].fillna(modeVal)

  else:
    #Fill with the mean
    meanVal = np.mean(dfNew[col].to_numpy()) #Find mean of column
    dfNew[col] = dfNew[col].fillna(meanVal)

dfNew.head()
```
